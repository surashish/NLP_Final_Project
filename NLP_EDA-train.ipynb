```python
def writeFile(name, data):
    fw = open(name, "w+")
    fw.write(data)
    fw.close()
    return
```

### Extracting sender and receiver

```python
def extract_from(data):
    
    tree = ET.ElementTree(ET.fromstring(data))
    root = tree.getroot()
    
    from_name = []
    
    for something in root.iter(tag='from'):
        temp_name = something.get('name')
        from_name.append(temp_name)
        
    return set(from_name)


def extract_to(data):
    
    tree = ET.ElementTree(ET.fromstring(data))
    root = tree.getroot()
    
    to_name = []
    
    for something in root.iter(tag='to'):
        temp_name = something.get('name')
        to_name.append(temp_name)
        
    return set(to_name)
```

```python
def extract_message(data):
    
    tree = ET.ElementTree(ET.fromstring(data))
    root = tree.getroot()
    
    message = []
    for each in root.iter(tag='message'):
        message.append(each)
    
    text_message = []
    
    for each in message:
        text_message.append(ET.tostring(each))
        
    return text_message
```

```python
def extract_thread_id(data):
    
    tree = ET.ElementTree(ET.fromstring(data))
    root = tree.getroot()
    
    thread_id = []
    
    for each in root.iter(tag='thread_id'):
        thread_id.append(each.text)
        
    return set(thread_id)
```

```python
def extract_cc(data):
    
    tree = ET.ElementTree(ET.fromstring(data))
    root = tree.getroot()
    
    cc = []
    
    for something in root.iter(tag='cc'):
        temp_name = something.get('name')
        cc.append(temp_name)
            
    return set(cc)
```
